```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("postgresql+psycopg2://postgres:123456789@localhost:5432/retail_sales_db")
```

```python
query_data = "SELECT * FROM orders;"
data = pd.read_sql_query(query_data, engine)
data.head(3)
```

```python
query_analisis = """
    SELECT 
        s.order_id,
        s.order_date,
        s.store_id,
        st.region,
        s.customer_id,
        c.segment,
        s.product_id,
        p.product_name,
        p.category,
        s.quantity,
        s.unit_price,
        s.discount,
        s.cost
    FROM orders s
    LEFT JOIN customer c
    ON c.customer_id=s.customer_id
    LEFT JOIN product p
    ON p.product_id=s.product_id
    LEFT JOIN store st
    ON st.store_id=s.store_id
    ORDER BY s.order_date;
"""

df = pd.read_sql_query(query_analisis, engine)
df
```

```python
df.to_csv('data.csv')
```
